## Start

Importer dhlab og pandas 

```python
import dhlab as dh
import pandas as pd

# noe kode er ikke ferdig bygd inn
import dhlab.api.dhlab_api as api
```

## Korpus av tekster

Alt starter med en mengde tekster. 

```python
corpus_definition = dh.Corpus(doctype="digibok", subject="mat", from_year=1960, to_year=2010, lang='nob', limit=500)
```

Alle korpus tilhører klasse for NB. Vi skal hente ut selve datarammen fra klassen (se [Python4abm - datarammer](https://mich123prem.github.io/python4abm/kapittel8.html#dataanalyse-med-pandas)) og jobbe videre med den.

```python
corpus_pandas = corpus_definition.frame
```

## Inspiser korpus

Bruk pandas kommandoer for å finne bøker, eller velge delmengder av bøker

```python
# Få en liste over kolonnene i korpusobjektene
corpus_pandas.columns
```

Begrens korpuset til å vise de mest essensielle dataene. Gjør listen så liten du vil, men ta med dhlabid og urn

```python
corpus_pandas[['dhlabid', 'urn', 'title', 'authors','city', 'year', 'publisher', 'langs', 'subjects', 'ddc',
       'genres', 'literaryform', 'doctype']].head(5)
```

Vi legger korpus inn under navnet corpus - og lager en kopi av den opprinnelige rammen

```python
corpus = corpus_pandas[['dhlabid', 'urn', 'title', 'authors','city', 'year', 'publisher', 'langs', 'subjects', 'ddc',
       'genres', 'literaryform', 'doctype']].copy()
```

```python
corpus.head(5)
```

Vi kan sikre oss entydig verdi av språk med hjelp av pandas filtrering 

Les om hvordan lage delmenger med pandas i [Python4abm - filtrering](https://mich123prem.github.io/python4abm/kapittel8.html#utdrag-av-radene-filtrering)


```python
corpus['antall_språk'] = corpus['langs'].apply(lambda x: len(x.split('/')))
```

```python
corpus[corpus['antall_språk'] == 1].head(5)
```

## Hente data

Når vi har korpuset kan vi hente ut data for inspeksjon og analyse.

Det er to typer konkordanser, en med fulltekstsyntaks (trunkering nærhetskrav), og en som matcher nøyaktig.

```python
conc = dh.Concordance(corpus=corpus, query="NEAR(salt pepper, 10)")
```

```python
conc.size
```

```python
conc.show(10)
```

## Hente ut et kollokasjonsobjekt

Kollokasjoner er ord som forekommer sammen med et søkeuttrykk, ord eller frase eller betingelse

```python
coll = api.concordance_counts(urns= list(corpus['urn']), words="NEAR(salt pepper, 5)", window=25, limit=1000)
```

```python
coll_salt = api.concordance_counts(urns= list(corpus['urn']), words="salt", window=25, limit=1000)
```

Resultatet er en dataramme - her ser vi på den sortert

```python
coll_salt.sort_values(by="freq", ascending=False).head(20)
```

# Med enkeltord

```python
coll_word = dh.Collocations(corpus=corpus, words =['salt','Salt'], before=0,after=24)
```

```python
coll_word = coll_word.frame
```

```python
coll_word.sort_values(by='counts', ascending=False).head(20)
```

# Måle forskjeller

Se også på normalisering i [Python4abm - text in pandas](https://mich123prem.github.io/python4abm/kapittel8.html#analyse-av-kolonner-med-tekst)

```python
coll_word['relative'] = coll_word['counts']*100/coll_word['counts'].sum()
```

```python
coll_word
```

# Sammenligne

Her velger vi en total

```python
tot = dh.totals(500000)
```

```python
tot['tot_rel'] = tot['freq']*100/tot['freq'].sum()
```

```python
data = coll_word.merge(tot, left_index=True, right_index=True, how="left").copy()[['counts','relative','tot_rel']]
```

# Nå kan vi sammenligne relative verdier

```python
data.loc[['.',',']]
```

Finner relevansmål, mål på forskjell

```python
data['relevans'] = data['relative']/data['tot_rel']
```

Sortere på relevans - jo høyere, jo høyrer tilknytning (spm: mellom hva?)

```python
data.sort_values(by="relevans", ascending=False).head(20)
```

Hva betyr relevans, hva skjer med vanlige ord, ord som kanskje ikke har så mye å bidra med?

```python
data.loc[['.',',','og','i','med','bland','er']]
```

Bearbeide med forskjellige paramatre

```python
data[data['counts']<= 3].sort_values(by='relevans', ascending=False).head(20)
```

```python
data[data['relevans']>= 500].sort_values(by='counts', ascending=False).head(20)
```
